```python
# Creating a new column based on existing column :

df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.head()
```

```python
# Drop a column in dataframe
df_tracks.drop("duration_ms",inplace+True,axis=1)
```

```python
# Creating a Heat map
corr_df = df_tracks.corr(method="pearson",numeric_only=True)
plt.figure(figsize=(18, 9))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".2f", cmap="inferno", linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)

plt.show()
```

```python
# Creating a new column from index "release_date"
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks['dates'] = pd.to_datetime(df_tracks['dates'])
years = df_tracks['dates'].dt.year
```

```python
df_tracks.head()
```

```python
# Number of songs released every year
sns.displot(years,discrete= True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
# Bar plot showing Year wise Average duration of songs
Total_dr= df_tracks.duration
fig_dims=(18,7)
fig,ax= plt.subplots(figsize=fig_dims)
fig= sns.barplot(x=years,y=Total_dr,ax=ax,errwidth= False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
# import another dataset
df_genre=pd.read_csv("C:\\Users\\Ravi\\Desktop\\SpotifyFeatures.csv")
df_genre.head()
```

```python
# Duration of in different Genres using Bar chart
plt.title("Duration of in different Genres")
sns.color_palette("rocket",as_cmap= True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")

```
