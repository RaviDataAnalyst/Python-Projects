```python
# Replacing Null values in TotalCharges column.

df['TotalCharges']= df['TotalCharges'].replace(" ",'0')
```

```python
# Changing the datatype of TotalCharges column.

df['TotalCharges'] = df['TotalCharges'].astype("float")
```

```python
# Mapping 0 and 1 value in SeniorCitizen column with Yes and No.

def conv(value):
    if value ==1:
        return "Yes"
    else:
        return "No"
    
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df['SeniorCitizen'].head()
```

```python
# Creating a count plot showing no of customers chrun.

plt.figure(figsize=(7, 5), facecolor='lightgrey')
ax = sns.countplot(x='Churn', data=df, palette='husl', edgecolor='white', linewidth=2)
ax.set_facecolor('black')
plt.bar_label(ax.containers[0], color='white')
plt.title('Churn Counts')
plt.xlabel('Churn Status')
plt.ylabel('Count')
plt.show()
```

```python
# Creating a pie plot to calculate % of churn customers.

gb = df.groupby('Churn').agg({'Churn': "count"})  # Chrun column is of string datatype.
plt.figure(figsize= (7, 5),facecolor='lightgrey')
colors = ['#ff9999',  '#99ff99']  # Light red, green
plt.pie(gb['Churn'],labels = gb.index, autopct = '%1.1f%%',colors = colors, wedgeprops=dict(edgecolor='black', linewidth=2)
)
plt.title("Percentage of Chured Customers", fontsize = 12)
```

```python
# From the given pie chart we can conclude that 26.54% of customers have churned out.
```

```python
# Creating a count plot showing chrun by Gender.

plt.figure(figsize=(7, 5), facecolor='lightgrey')
ax = sns.countplot(x='gender', data=df, palette='husl', edgecolor='white', linewidth=2, hue ="Churn")
ax.set_facecolor('black')
# Loop through each container and add labels
for labels in ax.containers:
    plt.bar_label(labels, color='white')
plt.title('Gender Counts based on Churn', fontsize = 12)
plt.xlabel('Gender Status')
plt.ylabel('Count')
plt.show()
```

```python
# Creating a count plot showing chrun by Senior Citizens.

plt.figure(figsize=(7, 5), facecolor='lightgrey')
ax = sns.countplot(x='SeniorCitizen', data=df, palette='husl', edgecolor='white', linewidth=2, hue ="Churn")
ax.set_facecolor('black')
for labels in ax.containers:
    plt.bar_label(labels, color='white')
plt.title(' Churn by Senior Citizen', fontsize = 12)
plt.xlabel('SeniorCitizen Status') 
plt.ylabel('Count')
plt.show()
```

```python
# Comparative a greater number of people in senior citizen category have churned .
```

```python
# Creating a Histogram showing chrun based on Tenure.

plt.figure(figsize=(7,5), facecolor='lightgrey')
ax = sns.histplot(data=df, x='tenure', bins=20, binwidth=5, stat='count', edgecolor='white',palette='husl',
                  linewidth=2, color='blue', alpha=0.7, hue='Churn', multiple='stack',kde=True,)
ax.set_facecolor('black')
plt.title('Churn of Customers by Tenure ', fontsize = 12)
plt.xlabel('Tenue in Months') 
plt.ylabel('Count of Customers')
plt.show()
```

```python
# Person who have used our services for long time have stayed and peoplewho have used our services for less months have churned.
```

```python
# Creating Subplots

columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid (you can change this)
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))  # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays)
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"],palette='husl',edgecolor='black')
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
#The majority of customers who do not churn tend to have 
#services like PhoneService, InternetService (particularly DSL), 
#and OnlineSecurity enabled. For services like OnlineBackup, 
#TechSupport, and StreamingTV, churn rates are noticeably 
#higher when these services are not used or are unavailable.
```
